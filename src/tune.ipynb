```python
from tqdm import tqdm

def beautify_tqdm(iterable, desc='', total=None, ncols=100, mininterval=0.1, **kwargs):
    """
    返回一个美化的 tqdm 进度条。

    :param iterable: 可迭代对象（如列表、range等）
    :param desc: 进度条前缀文本
    :param total: 进度条的总数量
    :param ncols: 进度条的宽度
    :param mininterval: 进度条更新的最小时间间隔
    :param kwargs: 其他参数传递给 tqdm
    :return: 返回美化后的 tqdm 进度条
    """
    # 使用 tqdm 初始化进度条
    bar = tqdm(iterable, desc=desc, total=total, ncols=ncols, mininterval=mininterval, **kwargs)

    # 设置进度条的颜色和样式
    bar.set_description(f'\033[1;32;40m{desc}\033[0m')  # 设置描述为绿色
    bar.set_bar_format(
        '{l_bar}{bar}| {n_fmt}/{total_fmt} [elapsed: {elapsed} remaining: {remaining}]'
    )  # 设置进度条格式

    # 自定义美化后的进度条样式
    bar.set_postfix_str('')  # 清除后缀
    return bar

# 使用示例
if __name__ == "__main__":
    for i in beautify_tqdm(range(100), desc="Processing", unit="item"):
        # 模拟任务
        pass
```
